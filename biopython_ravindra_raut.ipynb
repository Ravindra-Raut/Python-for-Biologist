```python
## Assinment 6-- write program to generate the quality control report of a given primer sequence- 
# there are two conditions for good primer
# 1) length of primer must be >18 bp
# 2) melting tm of primer should be in range of 55-65 degree centigrade
# tm = 4(G+C)+2(A+T)
# if primer statisfies both conditions -its good else bad 
```

```python
def calculate_tm(primer):
    gc_count = primer.count('G') + primer.count('C')
    at_count = primer.count('A') + primer.count('T')
    tm = 4 * gc_count + 2 * at_count
    print("tm of primer=",tm)
    return tm

def is_good_primer(primer):
    length = len(primer)
    tm = calculate_tm(primer)

    if length > 18 and 55 <= tm <= 65:
        return True
    else:
        return False


def generate_good_primer():
    primer = input("Enter a primer sequence: ")
    while not is_good_primer(primer):
        length = len(primer)
        tm = calculate_tm(primer)

        if length <= 18:
            print("Entered primer is too short. Please enter a longer seq")
        else:
            print("Entered primer has an inappropriate melting temperature. Please enter a new sequence")

        primer = input("Enter a primer sequence: ")

    return primer

good_primer = generate_good_primer()
print("Entered Good Primer=", good_primer)
```

```python
# import regular expression library 
import re 
```

```python
# search space
dna="AACGTTCGATGCCAGGAATCATGAACAACCGGGTATG"

# SEARCH pattern
pattern="ATG"
```

```python
# finding the FIRST match
if re.search(pattern,dna):
   result=re.search(pattern,dna)
   print(result.start()+1,result.end(),result.group(0))

else:
    print("no match found")
```

```python
## finding All matches in string
if re.search(pattern,dna):
 for i in re.finditer(pattern, dna):
  print(i.start()+1,i.end(),i.group(0))

else:
    print("no match found")
```
